```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
```

```python
X = np.array([[-1, 0], [-1, 1], [1, 0]])
y = np.array([-1, -1, 1])
```

```python
clf = svm.SVC(kernel='linear', C=1e10)
```

```python
clf.fit(X, y)
```

```python
w = clf.coef_[0]
b = clf.intercept_[0]
```

```python
plt.scatter(X[y == -1, 0], X[y == -1, 1], marker='o', color='black', label='Class O', s=100)
plt.scatter(X[y == 1, 0], X[y == 1, 1], marker='x', color='black', label='Class X', s=100)

if w[1] != 0:
    slope = -w[0] / w[1]
    intercept = -b / w[1]
    xx = np.linspace(-3, 3)
    yy = slope * xx + intercept
    plt.plot(xx, yy, 'k-', label='H')
else:
    plt.axvline(x=-b/w[0], color='k', linestyle='--', label='H')
    
plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=200, facecolors='none', edgecolors='k')
    


plt.title('Maximum Margin Hyperplane (H)')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.show()
```

```python
new_point = np.array([[2, 0]])
X_new = np.vstack([X, new_point])
y_new = np.array([-1, -1, 1, 1])
```

```python
clf_new = svm.SVC(kernel='linear', C=1e10)
clf_new.fit(X_new, y_new)
```

```python
w_new = clf_new.coef_[0]
b_new = clf_new.intercept_[0]
```

```python
plt.scatter(X_new[y_new == -1, 0], X_new[y_new == -1, 1], marker='o', color='black', label='Class O', s=100)
plt.scatter(X_new[y_new == 1, 0], X_new[y_new == 1, 1], marker='x', color='black', label='Class X', s=100)

if w_new[1] != 0:
    slope_new = -w_new[0] / w_new[1]
    intercept_new = -b_new / w_new[1]
    xx_new = np.linspace(-3, 3)
    yy_new = slope_new * xx_new + intercept_new
    plt.plot(xx_new, yy_new, 'r--', label='H\'')
else:
    plt.axvline(x=-b_new/w_new[0], color='r', linestyle='--', label='H\'')
    
plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=200, facecolors='none', edgecolors='k')
    
plt.title('Updated Data Points with Decision Boundary (H\')')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.show()
```
